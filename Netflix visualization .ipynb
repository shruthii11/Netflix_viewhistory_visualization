```python
monthly_counts = df['Month'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
monthly_counts.plot(kind='bar', color='skyblue')
plt.title('Netflix Viewing Frequency by Month')
plt.xlabel('Month')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
sns.set_style('whitegrid')

by_year = df.groupby('Month')['Year'].value_counts().sort_values(ascending=False).reset_index(name='Frequency')

# Create a bar plot
plt.figure(figsize=(15, 10))
plt.subplot(2, 1, 1)
ax = sns.barplot(x=by_year["Month"], y=by_year["Frequency"], hue=by_year["Year"], palette="cividis", alpha=0.7)
plt.title("Most Watched Month")
plt.xlabel("Month")
plt.ylabel("Frequency")

# Display the plot
plt.tight_layout()
plt.show()
```

```python

most_watched_per_month = df.groupby(['Month'])['show_name'].agg(lambda x: x.value_counts().index[0])

# Plot the most-watched show for each month
plt.figure(figsize=(12, 8))
most_watched_per_month.value_counts().plot(kind='bar', color='salmon')
plt.title('Most-Watched Show by Month')
plt.xlabel('Month')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))
movies_watched = df[df['show_type'] == 'Movie']
movies_watched_by_year = movies_watched.groupby('Year')['show_name'].count().reset_index()

ax1.scatter(movies_watched_by_year['Year'], movies_watched_by_year['show_name'], color='skyblue', marker='o')
ax1.set_title('Distribution of Movies Watched Over the Years')
ax1.set_xlabel('Year')
ax1.set_ylabel('Number of Movies Watched')
ax1.grid(True)

shows_watched = df[df['show_type'] == 'TV Show']

shows_watched_by_year = shows_watched.groupby('Year')['show_name'].count().reset_index()

ax2.scatter(shows_watched_by_year['Year'], shows_watched_by_year['show_name'], color='lightcoral', marker='o')
ax2.set_title('Distribution of TV Shows Watched Over the Years')
ax2.set_xlabel('Year')
ax2.set_ylabel('Number of TV Shows Watched')
ax2.grid(True)

plt.tight_layout()
plt.show()
```
